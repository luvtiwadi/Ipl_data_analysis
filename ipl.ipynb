```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
ohe.fit(df[['Player_Type']])

```

```python
oheData=ohe.transform(df[['Player_Type']])
```

```python
ohe.get_feature_names_out()
```

```python
df1=pd.DataFrame(oheData.toarray(),columns=ohe.get_feature_names_out())
```

```python
df1=df1.astype(int)
```

```python
df1
```

```python
df=pd.concat([df,df1],axis=1)
```
